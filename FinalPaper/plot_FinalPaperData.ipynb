```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import csv

%matplotlib inline
```

```python
import pandas as pd 
import seaborn as sns

pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
df = pd.read_csv("./Japan_GDP.csv", index_col=0, parse_dates=True)

df = df.dropna(axis=1)
df = df.divide(1000000)
ax=df.plot(title= 'Japan Yearly GDP in Millions of US Dollars',marker='o',xlim=pd.Timestamp('2000-01-01'),legend=False)
ax.set_xlabel("Year")
ax.set_ylabel("GDP")
plt.ticklabel_format(style='plain', axis='y')
plt.savefig('plot_Japan_GDP.pdf')



df = pd.read_csv("./NewOrleans_GDP.csv", index_col=0, parse_dates=True)

df = df.dropna(axis=1)
df = df.divide(1000000)
ax=df.plot(title= 'New Orleans Yearly GDP in Millions of US Dollars',marker='o',xlim=pd.Timestamp('2000-01-01'),legend=False)
ax.set_xlabel("Year")
ax.set_ylabel("GDP")
plt.ticklabel_format(style='plain', axis='y')
plt.savefig('plot_NewOrleans_GDP.pdf')
```

```python
df = pd.read_csv("./Japan_unemployment.csv", index_col=0, parse_dates=True)

df = df.dropna(axis=1)
ax=df.plot(title= 'Japan Monthly Unemployment Rate',xlim=pd.Timestamp('2000-01-01'),legend=False)
ax.set_xlabel("Date")
ax.set_ylabel("Unemployment Rate")
plt.ticklabel_format(style='plain', axis='y')
plt.savefig('plot_Japan_unemployment.pdf')



df = pd.read_csv("./NewOrleans_unemployment.csv", index_col=0, parse_dates=True)

df = df.dropna(axis=1)
ax=df.plot(title= 'New Orleans Monthly Unemployment Rate',xlim=pd.Timestamp('2000-01-01'),legend=False)
ax.set_xlabel("Date")
ax.set_ylabel("Unemployment Rate")
plt.ticklabel_format(style='plain', axis='y')
plt.savefig('plot_NewOrleans_unemployment.pdf')
```
